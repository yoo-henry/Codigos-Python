### 📌 Requisitos do Código


✅ Entrada de Dados:


Utilize input() para receber peso (kg) e altura (m).

Certifique-se de que os valores sejam números positivos (valide isso!).





✅ Cálculo do IMC:

Use a fórmula fornecida e armazene o resultado em uma variável.






✅ Classificação Condicional:

Utilize if, elif e else para determinar a categoria do IMC.





✅ Saída Formatada:

Mostre a classificação e uma recomendação personalizada (ex: "Procure um nutricionista" para obesidade).



✅ Tratamento de Erros:

Se o usuário digitar um valor inválido (negativo ou não numérico), exiba: "Entrada inválida. Digite apenas números positivos."

```python
### Ajudando Vocês

print("=== CALCULADORA DE IMC ===")
### Pedi o nome da pessoa. 
nome = input("Digite o seu nome")
### Aqui foi feita uma pergunta para saber o peso e a altura na pessoa.
peso = float(input(f"Digite o seu peso senhor(a) {nome} "))
altura = float(input(f"Digite a sua altura senhor(a){nome}"))
### Formula do IMC
imc = peso / (altura ** 2)


### Se for abaixo desse número a pessoa vai ser classificada como abaixa do peso.
if imc < 18.5:
    print(f"{nome}, o seu imc é: {round(imc,2)} a sua classificação é abaixo do peso, {nome} recomendo você ir comer algum lanche seu palito de dente.🥘")
 ### Se for maior ou menos dos números abaixo, a pessoa é classificada como normal.
elif imc >= 18.5 and imc <= 24.9:
    print(f"{nome}, o seu imc é: {round(imc,2)} a sua classificação é peso normal, {nome} parabéns você é uma pessoa normal.🎊")
### Se for maior ou menos dos números abaixo, a pessoa é classificada como Sobrepeso.
elif imc >= 25.0 and imc <= 29.9:
    print(f"{nome}, o seu imc é: {round(imc,2)} a sua classificação é Sobrepeso, recomendo diminuir um pouco na comida, fica ligado. 😉")
### Se for maior ou menos dos números abaixo, a pessoa é classificada como Obesidade Grau I.
elif imc >= 30.0 and imc <= 34.9:
    print(f"{nome}, o seu imc é: {round(imc,2)} a sua classificação é Obesidade Grau I, recomendo procurar algum médico, antes que piora. ⚠️")
### Se for maior ou menos dos números abaixo, a pessoa é classificada como Obesidade Grau II.
elif imc >= 35.0 and imc <= 39.9:
    print(f"{nome}, o seu imc é: {round(imc,2)} a sua classificação é Obesidade Grau II, recomendo passar em um médico antes que fica pior.🚨")
### E se nao for nenhum dos de cima, a pessoa é classificada como Obesidade Grau III.
else:
    print(f"{nome}, o seu imc é: {round(imc,2)} a sua classificação é Obesidade Grau III ⚠️🚨, recomendo você ir correndo para o médico mais proximo urgentemente.🏃‍➡️ ")

```
