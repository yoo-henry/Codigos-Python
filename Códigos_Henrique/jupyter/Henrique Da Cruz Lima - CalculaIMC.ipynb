```python
### Ajudando Voc√™s

print("=== CALCULADORA DE IMC ===")
### Pedi o nome da pessoa. 
nome = input("Digite o seu nome")
### Aqui foi feita uma pergunta para saber o peso e a altura na pessoa.
peso = float(input(f"Digite o seu peso senhor(a) {nome} "))
altura = float(input(f"Digite a sua altura senhor(a){nome}"))
### Formula do IMC
imc = peso / (altura ** 2)


### Se for abaixo desse n√∫mero a pessoa vai ser classificada como abaixa do peso.
if imc < 18.5:
    print(f"{nome}, o seu imc √©: {round(imc,2)} a sua classifica√ß√£o √© abaixo do peso, {nome} recomendo voc√™ ir comer algum lanche seu palito de dente.ü•ò")
 ### Se for maior ou menos dos n√∫meros abaixo, a pessoa √© classificada como normal.
elif imc >= 18.5 and imc <= 24.9:
    print(f"{nome}, o seu imc √©: {round(imc,2)} a sua classifica√ß√£o √© peso normal, {nome} parab√©ns voc√™ √© uma pessoa normal.üéä")
### Se for maior ou menos dos n√∫meros abaixo, a pessoa √© classificada como Sobrepeso.
elif imc >= 25.0 and imc <= 29.9:
    print(f"{nome}, o seu imc √©: {round(imc,2)} a sua classifica√ß√£o √© Sobrepeso, recomendo diminuir um pouco na comida, fica ligado. üòâ")
### Se for maior ou menos dos n√∫meros abaixo, a pessoa √© classificada como Obesidade Grau I.
elif imc >= 30.0 and imc <= 34.9:
    print(f"{nome}, o seu imc √©: {round(imc,2)} a sua classifica√ß√£o √© Obesidade Grau I, recomendo procurar algum m√©dico, antes que piora. ‚ö†Ô∏è")
### Se for maior ou menos dos n√∫meros abaixo, a pessoa √© classificada como Obesidade Grau II.
elif imc >= 35.0 and imc <= 39.9:
    print(f"{nome}, o seu imc √©: {round(imc,2)} a sua classifica√ß√£o √© Obesidade Grau II, recomendo passar em um m√©dico antes que fica pior.üö®")
### E se nao for nenhum dos de cima, a pessoa √© classificada como Obesidade Grau III.
else:
    print(f"{nome}, o seu imc √©: {round(imc,2)} a sua classifica√ß√£o √© Obesidade Grau III ‚ö†Ô∏èüö®, recomendo voc√™ ir correndo para o m√©dico mais proximo urgentemente.üèÉ‚Äç‚û°Ô∏è ")

```
