```python
def contar_vogais(texto):
    vogais = "aeiou"
    contador = 0

    texto = texto.lower()
    for i in texto:
        if i in vogais:
            contador += 1
    return contador

texto_usuario = input("Digite uma frase: ")
resultado = contar_vogais(texto_usuario)
print(f"A sua frase {texto_usuario}, tem {resultado} vogais.")
```

```python
def calcular_imc(peso, altura):
    imc = peso / (altura * altura)

    return imc

peso = float(input("Digite o seu peso: "))
altura = float(input("Digite a sua altura"))

resultado = calcular_imc(peso, altura)
print(f"O seu imc é {resultado}")
```
