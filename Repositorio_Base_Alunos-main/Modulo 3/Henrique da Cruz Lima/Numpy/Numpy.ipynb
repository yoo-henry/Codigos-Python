```python
import numpy as np
```

Array Numpy(Lista Otimizada)

```python
numeros = [2,4,8,16,32,64,128]

array_numeros = np.array(numeros)

array_numeros
```

<b> Operações Vetorizadas </b>

#numeros +2 --> Com uma lista normal da erro
for numero in numeros:
    print(numero * 2)

print(array_numeros * 2)

Comparações Vetorizadas


```python
array_aleatorios = np.array([22,12,13,14,54,23,25,75])
print(array_aleatorios % 2 == 0)
```

Filtro condicional

```python
print("Maiores que 20: ", array_aleatorios[array_aleatorios > 20]) # Mostra apenas os valores maiores que 20
print("Pares: ", array_aleatorios[array_aleatorios % 2 == 0]) # Mostra apenas os valores pares
print("Ímapares: ", array_aleatorios[array_aleatorios % 2 ==1]) # Mostra apenas os valores ímpares
```

<b> Operações com Arrays </b>

```python
print(array_aleatorios)
print("Soma",np.sum(array_aleatorios))
print("Média",np.mean(array_aleatorios))
print("Mínimo",np.min(array_aleatorios))
print("Máximo",np.max(array_aleatorios))
```
