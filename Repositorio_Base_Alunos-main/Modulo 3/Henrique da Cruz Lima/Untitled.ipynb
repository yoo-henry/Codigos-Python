```python
#maior >
#menos <
#igual ==
#diferente !=
#maior igual >=
temperatura = float(input("Digite o valor"))

if temperatura >=30:
    saida = "Muito quente!"
elif temperatura >= 24:
    saida = "Quente!"
elif temperatura >= 16:
    saida = "Agradavel!"
else:
    saida = "frio"

print(saida)
```

```python
#Conversão para números
#int = inteiro
#float = real

num1 = int(input("Digite o primeiro número: "))
num2 = int(input("Digite o segundo número: "))

print(f"A soma dos dois números são: {num1 + num2}")
print(f"A subtração dos dois números são: {num1 - num2}")
print(f"A multiplicação dos dois números são: {num1 * num2}")
print(f"A divisão dos dois números são: {num1 / num2}")
```

```python
nota1 = float(input("Digite a sua primeira nota: "))
nota2 = float(input("Digite a sua segunda nota: "))
nota3 = float(input("Digite a sua terceira nota: "))
media = (nota1 + nota2 + nota3) / 3

if media >= 7:
    saida = "Aprovado"
elif media >= 5:
    saida = "Recuperação!!"
else: 
    saida = "Reprovado!!"

print(saida)
```

```python
valor_compra = float(input("Digite o valor da sua compra: "))
valor_pago = float(input("Qual foi o valor pago ? "))

troco = valor_pago - valor_compra
falta = valor_compra - valor_pago

if valor_pago == valor_compra:
    saida = ("Está tudo certo, volte sempre!")
elif valor_pago > valor_compra:
    saida = (f"Seu troco é R${troco}")
elif valor_pago < valor_compra:
    saida = (f"Ainda está faltando R${falta}")

print(saida)
```
